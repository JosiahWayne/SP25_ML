```python
import numpy as np
import matplotlib.pyplot as plt

# This is similar to the code from lecture 1
# to sample from a 2D Gaussian Distribution

%config InlineBackend.figure_format = 'svg'
mean = [0,2]
cov = [[1, 2], [2, 5]]
# Fix random seed to get consistent results
np.random.seed(1024)
X = np.random.multivariate_normal(mean, cov, 1000)
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
fig.show()

plt.savefig('plot.pdf')
```

```python
# TODO: Report the estimate mean and covariance
# of the sampled points
mean_est = np.mean(X, axis=0)
cov_est = np.cov(X.T)
print("Estimated mean: ", mean_est)
print("Estimated covariance: ", cov_est)
```

```python
# Plot the histogram for the x-coordinates of X
plt.hist(X[:,0], bins=30, alpha=1, label='x-coordinates')
plt.legend()
plt.savefig('figure_x_coordinates.pdf')
plt.show()

# Plot the histogram for the y-coordinates of X
plt.hist(X[:,1], bins=30, alpha=1, label='y-coordinates')
plt.legend()
plt.savefig('figure_y_coordinates.pdf')
plt.show()
```

```python
# TODO: Are the x-coordinates of X samples from
# some Gaussian distribution?
# If so, estimate the mean and variance.
# Do the same for the y-coordinates.
miu_x = np.mean(X[:,0])
miu_y = np.mean(X[:,1])
var_x = np.var(X[:,0])
var_y = np.var(X[:,1])
print("Estimated mean of x-coordinates: ", miu_x)
print("Estimated variance of x-coordinates: ", var_x)
print("Estimated mean of y-coordinates: ", miu_y)
print("Estimated variance of y-coordinates: ", var_y)


# TODO: Generate a new 2D scatter plot of 1000 points,
# such that the x-coordinates(y-coordinates) of all the points
# are samples from a 1D Gaussian distribution 
# using the estimated mean and variance based on the x-coordinates(y-coordinates) of X.
mean = [miu_x, miu_y]
cov = [[var_x, 0], [0, var_y]]
# Fix random seed to get consistent results
np.random.seed(1024)
Y = np.random.multivariate_normal(mean, cov, 1000)
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(Y[:,0], Y[:,1], c='b', s=20.0, alpha=0.6, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
plt.savefig('plot2.pdf')
fig.show()
```

```python
# Back to the original X
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# TODO: Plot a line segment with x = [-10, 10]
# and y = 2x + 2 onto the 2D-Gaussian plot.
# The np.linspace() function may be helpful.
x = np.linspace(-10, 10, 100)
y = 2*x + 2
ax.plot(x, y, color='blue', label='y = 2x + 2')
plt.savefig('plot3.pdf')
fig.show()
```

```python
# TODO: Project X onto line y = 2x + 2
# and plot the projected points on the 2D space.
# You need to remove this line and assign the projected points to X_proj.
X_proj = np.zeros(X.shape)
for i in range(X.shape[0]):
    x, y = X[i]
    X_proj[i, 0] = (x + 2*y - 4) / 5
    X_proj[i, 1] = 2 * X_proj[i, 0] + 2


# Here we only plot 30 points to check the correctness
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X_proj[:30,0], X_proj[:30,1], c='b', s=40.0, alpha=0.6, label="Projected 2D-Gaussian")
ax.scatter(X[:30,0], X[:30,1], c='r', s=40.0, alpha=0.6, label="2D-Gaussian")
# Plot the project points:

ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
# You can also add the line from the previous plot for verification.
x = np.linspace(-10, 10, 100)
y = 2*x + 2
ax.plot(x, y, color='green', label='y = 2x + 2', alpha=0.5)
plt.savefig('figure4.5.pdf')
fig.show()
```
