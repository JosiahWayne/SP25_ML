```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
%config InlineBackend.figure_format = 'svg'
samples_X = np.arange(0, 20, 1) 
samples_Y = samples_X * 2 + np.random.normal(0, 1, len(samples_X))
regression_X = samples_X.reshape(-1, 1)
regression_Y = samples_Y.reshape(-1, 1)
regression = LinearRegression()
regression.fit(regression_X, regression_Y)
# draw the regression line 
plt.plot(samples_X, regression.predict(regression_X), color='red', label='Linear Regression')
ridge_regression = Ridge(alpha=150.0)
ridge_regression.fit(regression_X, regression_Y)
# draw the ridge regression line
plt.plot(samples_X, ridge_regression.predict(regression_X), color='green', label = "Ridge Regression")
plt.scatter(samples_X, samples_Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.savefig("ridge_regression.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
%config InlineBackend.figure_format = 'svg'
samples_X = np.arange(0, 20, 1) 
samples_Y = samples_X * 2 + np.random.normal(0, 1, len(samples_X))
samples_Y[14:18] = samples_Y[14:18] + 20
regression_X = samples_X.reshape(-1, 1)
regression_Y = samples_Y.reshape(-1, 1)
regression = LinearRegression()
regression.fit(regression_X, regression_Y)
# draw the regression line 
plt.plot(samples_X, regression.predict(regression_X), color='red', label='Linear Regression')
ridge_regression = Ridge(alpha=270.0)
ridge_regression.fit(regression_X, regression_Y)
# draw the ridge regression line
plt.plot(samples_X, ridge_regression.predict(regression_X), color='green', label = "Ridge Regression")
plt.scatter(samples_X, samples_Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.savefig("ridge_regression_f2.pdf")
plt.show()
```
